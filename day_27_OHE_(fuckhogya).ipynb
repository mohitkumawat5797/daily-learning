```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
'''
sex and embarked pe OHE lagaunga
'''
```

```python
df['Sex'].isnull().sum()
df['Embarked'].isnull().sum()
```

```python
'''
embarked me 2 missing values h ispe simple imputer lagaunga 2 value bharne ke liye phir uske baad lagega sex & embarked pe OHE
lekin ye sab krne se phle train test split bhi krna padega
'''
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'most_frequent')
df[['Embarked']] = si.fit_transform(df[['Embarked']])
```

```python
df['Embarked'].isnull().sum()
```

```python
'''ok so now null values ko most frquent values se bhar dia apan ne 
ab hogi 
OHE
'''
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop = 'first', sparse_output  = False)
encoded = ohe.fit_transform(df[['Sex','Embarked']])
```

```python
encoded
'''
ab is array ko df me convert karna padega
'''
```

```python
encoded_df = pd.DataFrame(encoded,columns = [['Sex','Embarked']])
encoded_df
#galat hai ye 
```
