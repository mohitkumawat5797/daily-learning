```python
'''
data pe encoding lagayenge

gender pe OHE
city me OHE
fever me value missing h unko fill karenge 
cough me Ordinal Encoding kyuki order h 

'''
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
```

```python
'''
phle simple tareeke se karke dekhenge
phir column transformer use krke
'''
```

```python
'''
simple imputer lagayenge fever ki missing values fill krne ke liye 
'''
from sklearn.impute import SimpleImputer
si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[['fever']])
X_test_fever = si.fit_transform(X_test[['fever']])
```

```python
X_train_fever.shape ,X_test_fever.shape 
```

```python
'''
Cough me ordinal encoding
'''
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = [['Mild','Strong']])
X_train_cough = oe.fit_transform(X_train[['cough']])
X_test_cough = oe.fit_transform(X_test[['cough']])
```

```python
X_train_cough.shape ,X_test_cough.shape
```

```python
'''
gender or city pe OHE
'''
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop = 'first' ,sparse_output = False)
X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])
X_test_gender_city = ohe.fit_transform(X_test[['gender','city']])
```

```python
X_train_gender_city
```

```python
'''
age col ko bhi ab as a array lena padega concatinate karne ke liye sabko
'''
X_train_age = X_train.drop(columns = ['gender','fever','cough','city']).values
X_test_age = X_test.drop(columns = ['gender','fever','cough','city']).values
X_train_age.shape
```

```python
#ab saare col ko jodna hai
X_train_final = np.concatenate((X_train_fever,X_train_cough , X_train_gender_city, X_train_age),axis = 1)
X_test_final = np.concatenate((X_test_fever,X_test_cough , X_test_gender_city, X_test_age),axis = 1)
X_train_final.shape
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```
