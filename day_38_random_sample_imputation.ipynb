```python
x = df.drop(columns = ['Survived'])
y = df['Survived']
```

```python
# isse kya kua na ki age col me jitne
# bhi na the unke liye sample random values apan ne bana li
# lekin ab un na vlaues ko ye ranodm val assign karni hai



random_ages = X_train['Age'].dropna().sample(X_train['Age'].isnull().sum()).values
```

```python
nan_indexes = X_train['Age'][X_train['Age'].isnull()].index
```

```python
X_train.loc[nan_indexes , 'Age'] = random_ages
```
