```python
# =================== 1. Data dekhna ===================
df.head()          # top 5 rows
df.head(10)        # top 10 rows
df.tail()          # last 5 rows
df.shape           # (rows, columns)
df.info()          # summary of columns + datatypes
df.columns         # list of column names
df.describe()      # numeric summary stats

# =================== 2. Specific columns dekhna ===================
df['col']              # single column
df[['col1','col2']]    # multiple columns
df.dtypes              # datatypes of all columns

# =================== 3. Specific rows dekhna ===================
df.iloc[0]                      # first row (position-based)
df.iloc[0:5]                    # first 5 rows
df.loc[3]                       # row with index label 3
df.loc[3:6, ['col1','col2']]    # specific rows + specific columns

# =================== 4. Filtering data ===================
df[df['Age'] > 25]                             # filter by condition
df[df['City'] == 'Delhi']                      # filter by category
df[(df['Age'] > 25) & (df['Salary'] < 50000)]  # multiple conditions

# =================== 5. Unique values check karna ===================
df['col'].unique()        # unique values in a column
df['col'].nunique()       # number of unique values
df['col'].value_counts()  # frequency of each unique value

# =================== 6. Chhote changes karna ===================
df.rename(columns={'old':'new'}, inplace=True)   # rename column
df['newcol'] = df['col1'] + df['col2']           # add new column
df.drop('col', axis=1, inplace=True)             # drop column
df.drop(0, axis=0, inplace=True)                 # drop row
df['col'] = df['col'].fillna(0)                  # fill missing values

# =================== 7. Sorting ===================
df.sort_values(by='Age')                          # ascending sort
df.sort_values(by='Salary', ascending=False)      # descending sort

# =================== 8. Grouping & Aggregation ===================
df.groupby('City')['Salary'].mean()   # average salary by city
df.groupby('ClassLabel').size()       # count of rows per class

# =================== 9. Indexing / Resetting ===================
df.reset_index(drop=True, inplace=True)   # reset index
df.set_index('col', inplace=True)         # set column as index
```

```python
# =================== 10. Describe aur Summary ===================
df.describe()                     # summary of numeric columns
df.describe(include='all')        # summary of all columns (numeric + categorical)
df['col'].describe()              # summary of single column

# =================== 11. Missing Data Check ===================
df.isnull().sum()                 # missing values per column
df.isnull().any()                 # check if any null present
df.notnull().sum()                # non-null count
df.dropna(inplace=True)           # drop rows with nulls
df.fillna(0, inplace=True)        # replace nulls with 0
df.fillna(df['col'].mean(), inplace=True)   # fill nulls with mean

# =================== 12. Value Replacement ===================
df['col'].replace({'old':'new'}, inplace=True)   # replace values in a column
df['col'] = df['col'].map({'Low':0, 'Medium':1, 'High':2})  # map categorical to numbers

# =================== 13. Duplicates ===================
df.duplicated().sum()             # count duplicates
df.drop_duplicates(inplace=True)  # remove duplicates

# =================== 14. Sampling ===================
df.sample(5)                      # random 5 rows
df.sample(frac=0.1)               # 10% random rows

# =================== 15. Correlation & Covariance ===================
df.corr()                         # correlation between numeric columns
df.cov()                          # covariance matrix

# =================== 16. Sorting by Index ===================
df.sort_index()                   # sort rows by index
df.sort_index(axis=1)             # sort columns by name

# =================== 17. Apply Function ===================
df['col'].apply(lambda x: x*2)    # apply function to a column
df.applymap(str)                  # apply function to whole dataframe

# =================== 18. Pivot Tables ===================
df.pivot_table(values='Salary', index='City', aggfunc='mean')   # pivot by city

# =================== 19. Crosstab (Frequency Table) ===================
pd.crosstab(df['City'], df['ClassLabel'])    # count combinations

# =================== 20. Export / Import ===================
df.to_csv('output.csv', index=False)         # save as CSV
pd.read_csv('output.csv')                    # read CSV
df.to_excel('output.xlsx', index=False)      # save as Excel
pd.read_excel('output.xlsx')                 # read Excel
```
