```python
import pandas as pd
import seaborn as sns
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(data = tips , x = 'total_bill'  , y = 'tip'   ,  hue  = 'sex' , style = 'smoker' , size = 'size' )
```

```python
sns.relplot(data = tips, kind = 'scatter' , x = 'total_bill'  , y = 'tip'   ,  hue  = 'sex' , style = 'smoker' , size = 'size' , col = 'time' )
```
